### 🔍 Диагностика моделей
Проверим, корректно ли созданы симлинки и какие файлы реально видны приложению.

```python
import os

def check_links(path):
    print(f"Checking models in: {path}")
    for f in os.listdir(path):
        full_p = os.path.join(path, f)
        is_link = os.path.islink(full_p)
        exists = os.path.exists(full_p)
        status = "✅ OK" if exists else "❌ BROKEN LINK"
        print(f"[{'Link' if is_link else 'File'}] {f} -> {status}")

check_links('/content/my_models/')
```
